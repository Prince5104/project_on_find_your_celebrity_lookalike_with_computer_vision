```python
import os
import numpy as np
from IPython.display import Image 
```

- <font size=3>**Create Function to Load Images and Face Encodings**</font>

```python
def load_images(known_image_dir):
    known_encodings = []
    known_image = []
    
    for file in os.listdir(known_image_dir):
        
        filename = os.fsdecode(file)
        image = face_recognition.load_image_file(os.path.join(known_image_dir, filename))
        
        enc = face_recognition.face_encodings(image)
        
        if len(enc) > 0:
            known_encodings.append(enc[0])
            known_image.append(filename)
            
    return (known_encodings, known_image)
```

- <font size=3>**Calculate Euclidean Distance**</font>

```python
# Create a function named calculate_face_distance() as given below:
def calculate_face_distance(known_encodings, unknown_img_path, cutoff=0.5, num_results=4):
    image_to_test = face_recognition.load_image_file(unknown_img_path)
    image_to_test_encoding = face_recognition.face_encodings(image_to_test)[0]

    face_distances = face_recognition.face_distance(known_encodings, image_to_test_encoding)
    return (unknown_img_path, known_images[face_distances.argmin()])
```

```python
# First, we load the celebrity images using the load_images() function:
known_encodings, known_images = load_images("/cxldata/projects/lookalikeceleb/images")
```

- <font size=3>**Find the Matching Celebrity Image**</font>

```python
# Use the calculate_face_distance() function to calculate the Euclidean distance between the faces:
matching_image =calculate_face_distance(known_encodings, original_image)[1]
```
